```python
import os
import cv2
import numpy as np
from keras.models import load_model

# Load the trained model
model = load_model('new_accident_model20april.h5')

# Constants
IMAGE_HEIGHT, IMAGE_WIDTH = 64, 64
SEQUENCE_LENGTH = 16
CLASSES_LIST = ["NonViolence", "Violence"]

# Function to preprocess frame and predict anomaly
def preprocess_frame(frame):
    resized_frame = cv2.resize(frame, (IMAGE_HEIGHT, IMAGE_WIDTH))
    return resized_frame / 255.0

# Function to play video and predict frame-wise labels
def play_video_with_predictions(video_file_path):
    video_reader = cv2.VideoCapture(video_file_path)
    if not video_reader.isOpened():
        print("Error: Could not open video.")
        return

    frame_sequence = []

    while True:
        ret, frame = video_reader.read()
        if not ret:
            break

        # Preprocess the frame
        preprocessed_frame = preprocess_frame(frame)
        frame_sequence.append(preprocessed_frame)

        # If the sequence length is reached, predict anomaly
        if len(frame_sequence) == SEQUENCE_LENGTH:
            # Convert the sequence into a numpy array
            sequence_array = np.array(frame_sequence)

            # Predict anomaly for the sequence
            predicted_labels_probabilities = model.predict(np.expand_dims(sequence_array, axis=0))[0]
            predicted_label_index = np.argmax(predicted_labels_probabilities)
            predicted_class = CLASSES_LIST[predicted_label_index]

            # Display the frame with predicted label
            cv2.putText(frame, predicted_class, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
            cv2.imshow('Video', frame)

            # Check if 'q' key is pressed to exit
            if cv2.waitKey(25) & 0xFF == ord('q'):
                break

            # Remove the oldest frame from the sequence
            frame_sequence.pop(0)

    video_reader.release()
    cv2.destroyAllWindows()

# Specify the path to the random input video
video_path = "videos/V_14.mp4"

# Play the video with frame-wise predictions
play_video_with_predictions(video_path)
```
